```python
# Interpolation Search
def interpolation_search(arr, key):
    low = 0
    high = len(arr) - 1
    probes = 0

    while low <= high and key >= arr[low] and key <= arr[high]:
        probes += 1

        if low == high:
            if arr[low] == key:
                return low, probes
            return -1, probes

        pos = low + int(
            ((high - low) / (arr[high] - arr[low]))
            * (key - arr[low])
        )

        if arr[pos] == key:
            return pos, probes

        elif arr[pos] < key:
            low = pos + 1

        else:
            high = pos - 1

    return -1, probes


# Binary Search
def binary_search(arr, key):
    low = 0
    high = len(arr) - 1
    probes = 0

    while low <= high:
        probes += 1

        mid = (low + high) // 2

        if arr[mid] == key:
            return mid, probes

        elif arr[mid] < key:
            low = mid + 1

        else:
            high = mid - 1

    return -1, probes


# Main Program
n = int(input("Enter number of students: "))

print("Enter sorted roll numbers:")
arr = list(map(int, input().split()))

key = int(input("Enter roll number to search: "))

interp_index, interp_probes = interpolation_search(arr, key)
binary_index, binary_probes = binary_search(arr, key)

print("\n--- Interpolation Search ---")
if interp_index != -1:
    print(f"Roll number found at index {interp_index}")
else:
    print("Roll number not found")
print("Probes required:", interp_probes)

print("\n--- Binary Search ---")
if binary_index != -1:
    print(f"Roll number found at index {binary_index}")
else:
    print("Roll number not found")
print("Probes required:", binary_probes)

print("\n--- Comparison ---")
if interp_probes < binary_probes:
    print("Interpolation Search used fewer probes.")
elif interp_probes > binary_probes:
    print("Binary Search used fewer probes.")
else:
    print("Both used the same number of probes.")
```
