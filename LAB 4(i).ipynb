```python
import heapq

def dijkstra_all_paths(graph, start):
    n = len(graph)

    # Distances initialized to infinity
    distances = [float('inf')] * n
    distances[start] = 0

    # Priority queue: (distance, node)
    pq = [(0, start)]

    # To reconstruct paths
    previous = [None] * n

    while pq:
        current_dist, u = heapq.heappop(pq)

        if current_dist > distances[u]:
            continue

        for v, weight in graph[u]:
            distance = current_dist + weight
            if distance < distances[v]:
                distances[v] = distance
                previous[v] = u
                heapq.heappush(pq, (distance, v))

    # Reconstruct paths for all nodes
    paths = {}
    for target in range(n):
        if distances[target] == float('inf'):
            paths[target] = None
        else:
            path = []
            node = target
            while node is not None:
                path.append(node)
                node = previous[node]
            path.reverse()
            paths[target] = path

    return distances, paths


# Example directed weighted graph with 6 vertices (0–5)
graph = {
    0: [(1, 7), (2, 9), (5, 14)],
    1: [(2, 10), (3, 15)],
    2: [(3, 11), (5, 2)],
    3: [(4, 6)],
    4: [(5, 9)],
    5: []
}

# Run Dijkstra from source vertex 0
distances, paths = dijkstra_all_paths(graph, 0)

# Print results
for v in range(len(graph)):
    if distances[v] == float('inf'):
        print(f"Vertex {v}: No path from 0")
    else:
        print(f"Vertex {v}: Distance = {distances[v]}, Path = {' -> '.join(map(str, paths[v]))}")
```
