```python
import random

def prim_matrix(graph, V):
    selected = [False] * V
    selected[0] = True
    mst_edges = []
    total_weight = 0

    for _ in range(V - 1):
        min_edge = (None, None, float('inf'))
        for u in range(V):
            if selected[u]:
                for v in range(V):
                    if not selected[v] and graph[u][v] != 0:
                        if graph[u][v] < min_edge[2]:
                            min_edge = (u, v, graph[u][v])
        u, v, w = min_edge
        mst_edges.append((u, v, w))
        total_weight += w
        selected[v] = True

    return mst_edges, total_weight

# Example: 6-city graph adjacency matrix
graph = [
    [0, 4, 2, 0, 0, 0],  # A
    [4, 0, 5, 10, 0, 0], # B
    [2, 5, 0, 0, 3, 0],  # C
    [0, 10, 0, 0, 7, 8], # D
    [0, 0, 3, 7, 0, 6],  # E
    [0, 0, 0, 8, 6, 0]   # F
]

mst, weight = prim_matrix(graph, 6)
print("MST edges (matrix):", mst)
print("Total weight:", weight)
```
