```python
import heapq

def dijkstra_with_hops(graph, start, target):
    n = len(graph)

    # Distances initialized to infinity
    distances = [float('inf')] * n
    hops = [float('inf')] * n
    distances[start] = 0
    hops[start] = 0

    # Priority queue: (distance, hops, node)
    pq = [(0, 0, start)]

    # To reconstruct path
    previous = [None] * n

    while pq:
        current_dist, current_hops, u = heapq.heappop(pq)

        # Skip if we already found a better path
        if (current_dist > distances[u]) or (current_dist == distances[u] and current_hops > hops[u]):
            continue

        for v, weight in graph[u]:
            distance = current_dist + weight
            hop_count = current_hops + 1

            # Update if shorter distance OR same distance but fewer hops
            if (distance < distances[v]) or (distance == distances[v] and hop_count < hops[v]):
                distances[v] = distance
                hops[v] = hop_count
                previous[v] = u
                heapq.heappush(pq, (distance, hop_count, v))

    # Reconstruct shortest path
    path = []
    node = target
    while node is not None:
        path.append(node)
        node = previous[node]
    path.reverse()

    return distances[target], hops[target], path


# Example graph with 6 vertices
# graph[node] = [(neighbor, travel_time), ...]
graph = {
    0: [(1, 7), (2, 9), (5, 14)],   # Node 0
    1: [(2, 10), (3, 15)],          # Node 1
    2: [(3, 11), (5, 2)],           # Node 2
    3: [(4, 6)],                    # Node 3
    4: [(5, 9)],                    # Node 4
    5: []                           # Node 5
}

# Find fastest route from node 0 to node 4
time, hop_count, route = dijkstra_with_hops(graph, 0, 4)

print("Fastest travel time:", time, "minutes")
print("Hop count:", hop_count)
print("Route:", " -> ".join(map(str, route)))
```
